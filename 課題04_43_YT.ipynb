# **YRPG**

## **概要**

このプログラムは魔王を倒すことを目的としたRPGです。プレイヤーはそれぞれの選択、そして成長を楽しむために自由度が高くなるようにプログラムを製作しました。
そのため、変数を活用し、様々な要素を盛り込んでいます。

## **操作方法**

imput()を多用しています。その際、なんと入力すればよいかは記載があるのでそちらをご覧ください。記載がない場合は何も入力しなくてかまいません。

## **制作者のコメント**

冒険者ギルドやアイテム売り場、鍛錬など様々な要素を盛り込んでみたので、そこに注目しながらプレイしてくださるとうれしいです。
プログラムでは、プレイヤーの選択肢を変数で管理し、次の選択肢を表示させていく部分を工夫しました。踏み込んだゲームの製作は初めてだったため、アイテムの処理の部分には苦労させられました。
